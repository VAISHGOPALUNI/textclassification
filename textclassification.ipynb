```python
import pandas as pd

```

```python
import numpy as np
```

```python
df = pd.read_csv("C:/Users/VAISHNAVI GOPALUNI/Downloads/projectdataset.csv")
df
```

```python
df.dropna()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], random_state=1)
X_train, X_test, y_train, y_test 

```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(strip_accents='ascii', token_pattern=u'(?ui)\\b\\w*[a-z]+\\w*\\b',lowercase=True, stop_words='english')
cv
```

```python
X_train_cv = cv.fit_transform(X_train)
X_train_cv
```

```python
X_test_cv = cv.transform(X_test)
X_test_cv
```

```python
word_freq_df = pd.DataFrame(X_train_cv.toarray(), columns=cv.get_feature_names())
word_freq_df
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
naive_bayes = MultinomialNB()
naive_bayes
```

```python
naive_bayes.fit(X_train_cv, y_train)
```

```python
predictions = naive_bayes.predict(X_test_cv)
predictions
```

```python
testing_predictions = []
for i in range(len(X_test)):
    if predictions[i] == '__label__1 ':
        testing_predictions.append('__label__1')
    else:
        testing_predictions.append('__label__2')
check_df = pd.DataFrame({'actual_label': list(y_test), 'prediction': testing_predictions, 'abstract':list(X_test)})
check_df
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Accuracy score:',accuracy_score(y_test, predictions))

```
